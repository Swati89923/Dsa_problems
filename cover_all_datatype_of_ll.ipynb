```python
# Node class - yeh ek node banata hai jisme data aur next pointer hota hai
class Node:
    def __init__(self, item=None, next=None):
        self.item = item      # item ka matlab hai node mein store kiya gaya data
        self.next = next      # next ek pointer hai jo agle node ko point karta hai

# CLL class - yeh Circular Linked List manage karti hai
class CLL:
    def __init__(self, last=None):
        self.last = last      # last pointer circular list ke last node ko point karta hai

    # Check karta hai ki list khaali hai ya nahi
    def is_empty(self):
        return self.last == None

    # Start mein naya node insert karta hai
    def insert_at_start(self, data):
        n = Node(data)  # naya node banaya
        if self.is_empty():  # agar list empty hai
            n.next = n       # khud ko hi point karega (circular banane ke liye)
            self.last = n    # wahi node ab last bhi ban gaya
        else:
            n.next = self.last.next     # naya node ab pehle wale first node ko point karega
            self.last.next = n          # last node ab naye node ko point karega (wo ab first node hai)

    # Last mein naya node insert karta hai
    def insert_at_last(self, data):
        n = Node(data)
        if self.is_empty():
            n.next = n
            self.last = n
        else:
            n.next = self.last.next     # naya node first node ko point karega
            self.last.next = n          # last node naye node ko point karega
            self.last = n               # naya node ab last ban gaya

    # Kisi particular data ko search karta hai
    def search(self, data):
        if self.is_empty():
            return None
        temp = self.last.next  # traversal start hoti hai first node se
        while temp != self.last:
            if temp.item == data:
                return temp     # agar data mil gaya to return kar do
            temp = temp.next
        if temp.item == data:
            return temp         # agar last node mein data ho to bhi check karo
        return None             # nahi mila to None return karo

    # Kisi particular node ke baad new node insert karna
    def insert_after(self, temp, data):
        if temp is not None:
            n = Node(data, temp.next)  # naya node bana aur temp ke baad insert kiya
            temp.next = n
            if temp == self.last:
                self.last = n          # agar last ke baad insert kiya, to last ko update karo

    # Poore list ko print karta hai
    def print_list(self):
        if not self.is_empty():
            temp = self.last.next  # first node se start karo
            while temp != self.last:
                print(temp.item, end=' ')
                temp = temp.next
            print(temp.item)  # last node print karo

    # First node ko delete karta hai
    def delete_first(self):
        if not self.is_empty():
            if self.last.next == self.last:
                self.last = None  # agar sirf ek node hai to list empty ho jaayegi
            else:
                self.last.next = self.last.next.next  # first node ko hata diya

    # Last node ko delete karta hai
    def delete_last(self):
        if not self.is_empty():
            if self.last.next == self.last:
                self.last = None
            else:
                temp = self.last.next  # first node se start karo
                while temp.next != self.last:
                    temp = temp.next
                temp.next = self.last.next  # last node hata diya
                self.last = temp            # naya last ban gaya

    # Kisi particular value/data wala node delete karna
    def delete_item(self, data):
        if not self.is_empty():
            # agar sirf ek hi node hai
            if self.last.next == self.last:
                if self.last.item == data:
                    self.last = None
            else:
                # agar first node delete karna hai
                if self.last.next.item == data:
                    self.delete_first()
                else:
                    temp = self.last.next
                    while temp != self.last:
                        if temp.next == self.last:
                            # agar last node delete karna hai
                            if self.last.item == data:
                                self.delete_last()
                            break
                        if temp.next.item == data:
                            temp.next = temp.next.next  # middle node ko skip kar diya
                            break
                        temp = temp.next

    # Yeh list ko iterable banata hai (for-loop ke liye)
    def __iter__(self):
        if self.last == None:
            return CLLIterator(None)
        else:
            return CLLIterator(self.last.next)

# Iterator class - list ko for-loop mein iterate karne ke liye
class CLLIterator:
    def __init__(self, start):
        self.current = start
        self.start = start
        self.count = 0

# --- Ab hum CLL class ka use karke circular linked list banayenge ---

cll = CLL()  # ek khaali circular linked list banayi

# Start mein nodes insert kar rahe hain
cll.insert_at_start(10)  # List: 10
cll.insert_at_start(20)  # List: 20 -> 10

# End mein nodes insert kar rahe hain
cll.insert_at_last(30)   # List: 20 -> 10 -> 30
cll.insert_at_last(40)   # List: 20 -> 10 -> 30 -> 40

# List ko print karte hain
cll.print_list()          # Output: 20 10 30 40
```

```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class CLL:
    def __init__(self):
        self.start = None

    def insert_first(self, data):
        nd = Node(data)
        if self.start is None:
            self.start = nd
            nd.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = nd
            nd.next = self.start
            self.start = nd

    def insert_last(self, data):
        nd = Node(data)
        if self.start is None:
            self.start = nd
            nd.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = nd
            nd.next = self.start

    def insert_after_loc(self, targeted, data):
        if self.start is None:
            print("List is empty!")
            return
        temp = self.start
        while True:
            if temp.data == targeted:
                nd = Node(data, temp.next)
                temp.next = nd
                return
            temp = temp.next
            if temp == self.start:
                print(f"Node with data {targeted} not found.")
                return

    def delete_first(self):
        if self.start is None:
            print("List is empty!")
            return
        if self.start.next == self.start:  # Only one node
            self.start = None
            return
        temp = self.start
        while temp.next != self.start:
            temp = temp.next
        temp.next = self.start.next
        self.start = self.start.next

    def delete_last(self):
        if self.start is None:
            print("List is empty!")
            return
        if self.start.next == self.start:  # Only one node
            self.start = None
            return
        temp = self.start
        prev = None
        while temp.next != self.start:
            prev = temp
            temp = temp.next
        prev.next = self.start

    def delete_after_loc(self, targeted):
        if self.start is None:
            print("List is empty!")
            return
        temp = self.start
        while True:
            if temp.data == targeted:
                if temp.next == self.start:  # deleting start node after last node
                    self.delete_first()
                else:
                    temp.next = temp.next.next
                return
            temp = temp.next
            if temp == self.start:
                print(f"Node with data {targeted} not found.")
                return

    def display(self):
        if self.start is None:
            print("List is empty!")
            return
        temp = self.start
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.start:
                break
        print(self.start.data)

# Example usage
cll = CLL()
cll.insert_first(10)
cll.insert_last(20)
cll.insert_last(30)
cll.insert_after_loc(20, 25)
cll.display()

print("\nDeleting first node...")
cll.delete_first()
cll.display()

print("\nDeleting last node...")
cll.delete_last()
cll.display()

print("\nDeleting after 20...")
cll.delete_after_loc(20)
cll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CLL:
    def __init__(self):
        self.start = None

    def insert_at_start(self, data):
        new_node = Node(data)
        if self.start is None:
            self.start = new_node
            new_node.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.start
            self.start = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.start is None:
            self.start = new_node
            new_node.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.start

    def insert_at_position(self, pos, data):
        if pos <= 0:
            print("Invalid position")
            return
        if pos == 1:
            self.insert_at_start(data)
            return
        new_node = Node(data)
        temp = self.start
        count = 1
        while count < pos - 1 and temp.next != self.start:
            temp = temp.next
            count += 1
        if count == pos - 1:
            new_node.next = temp.next
            temp.next = new_node
        else:
            print("Position out of bounds")

    def delete_at_start(self):
        if self.start is None:
            print("List is empty")
            return
        if self.start.next == self.start:
            self.start = None
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = self.start.next
            self.start = self.start.next

    def delete_at_end(self):
        if self.start is None:
            print("List is empty")
            return
        if self.start.next == self.start:
            self.start = None
        else:
            prev = None
            temp = self.start
            while temp.next != self.start:
                prev = temp
                temp = temp.next
            prev.next = self.start

    def delete_by_value(self, value):
        if self.start is None:
            print("List is empty")
            return
        if self.start.data == value:
            self.delete_at_start()
            return
        prev = self.start
        curr = self.start.next
        while curr != self.start:
            if curr.data == value:
                prev.next = curr.next
                return
            prev = curr
            curr = curr.next
        print("Value not found")

    def search(self, value):
        if self.start is None:
            print("List is empty")
            return False
        temp = self.start
        while True:
            if temp.data == value:
                return True
            temp = temp.next
            if temp == self.start:
                break
        return False

    def display(self):
        if self.start is None:
            print("List is empty")
            return
        temp = self.start
        while True:
            print(temp.data, end=" ")
            temp = temp.next
            if temp == self.start:
                break
        print()
cll = CLL()

# Inserting nodes
cll.insert_at_end(10)
cll.insert_at_end(20)
cll.insert_at_start(5)
cll.insert_at_position(2, 15)
cll.display()  # Output: 5 15 10 20

# Deleting nodes
cll.delete_by_value(15)
cll.delete_at_start()
cll.delete_at_end()
cll.display()  # Output: 10

# Searching
print("Found 10:", cll.search(10))  # True
print("Found 99:", cll.search(99))  # False
```

```python
class Node:
    def __init__(self, item=None, next=None):
        self.item = item
        self.next = next

class Cll:
    def __init__(self):
        self.start = None
        self.size = 0

    def insert_start(self, data):
        nd = Node(data)
        if self.start is None:
            self.start = nd
            nd.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = nd
            nd.next = self.start
            self.start = nd
        self.size += 1

    def insert_at_last(self, data):
        if self.start is None:
            self.insert_start(data)
            return
        nd = Node(data)
        temp = self.start
        while temp.next != self.start:
            temp = temp.next
        temp.next = nd
        nd.next = self.start
        self.size += 1

    def insert_pos(self, pos, data):
        if pos <= 0 or pos > self.size + 1:
            print("Invalid index")
            return
        if pos == 1:
            self.insert_start(data)
            return
        nd = Node(data)
        temp = self.start
        for _ in range(pos - 2):
            temp = temp.next
        nd.next = temp.next
        temp.next = nd
        self.size += 1

    def delete_first(self):
        if self.start is None:
            print("List is empty")
            return
        if self.start.next == self.start:
            self.start = None
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = self.start.next
            self.start = self.start.next
        self.size -= 1

    def delete_last(self):
        if self.start is None:
            print("List is empty")
            return
        if self.start.next == self.start:
            self.start = None
        else:
            prev = None
            temp = self.start
            while temp.next != self.start:
                prev = temp
                temp = temp.next
            prev.next = self.start
        self.size -= 1

    def delete_val(self, value):
        if self.start is None:
            print("List is empty")
            return
        if self.start.item == value:
            self.delete_first()
            return
        prev = self.start
        curr = self.start.next
        while curr != self.start:
            if curr.item == value:
                prev.next = curr.next
                self.size -= 1
                return
            prev = curr
            curr = curr.next
        print("Value not found")

    def display(self):
        if self.start is None:
            print("List is empty")
            return
        temp = self.start
        while True:
            print(temp.item, end=" ")
            temp = temp.next
            if temp == self.start:
                break
        print()
cl = Cll()
cl.insert_start(10)
cl.insert_start(20)
cl.insert_at_last(5)
cl.insert_pos(2, 15)
cl.display()           # Output: 20 15 10 5
cl.delete_first()
cl.delete_last()
cl.display()           # Output: 15 10
cl.delete_val(15)
cl.display()           # Output: 10
```
