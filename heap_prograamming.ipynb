```python
class EmptyHeapException(Exception):
    def __init__(self, msg="Empty heap"):
        self.msg = msg

    def __str__(self):
        return self.msg


class Heap:
    def __init__(self):
        self.heap = []

    # Corrected method: use self.insert(e), not list1.insert(e)
    def create_heap(self, list1):
        for e in list1:
            self.insert(e)

    def insert(self, e):
        index = len(self.heap)
        self.heap.append(e)
        parent_index = (index - 1) // 2

        # MISTAKE FIXED: You must update index and parent_index inside the loop
        while index > 0 and self.heap[parent_index] < self.heap[index]:
            self.heap[parent_index], self.heap[index] = self.heap[index], self.heap[parent_index]
            index = parent_index
            parent_index = (index - 1) // 2

    def top(self):
        if len(self.heap) == 0:
            raise EmptyHeapException()
        else:
            return self.heap[0]

    def delete_element(self):
        if len(self.heap) == 0:
            raise EmptyHeapException()

        max_value = self.heap[0]
        temp = self.heap.pop()

        if len(self.heap) == 0:
            return max_value

        self.heap[0] = temp
        index = 0
        size = len(self.heap)

        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            largest = index

            if left < size and self.heap[largest] < self.heap[left]:
                largest = left
            if right < size and self.heap[largest] < self.heap[right]:
                largest = right

            if index == largest:
                break

            self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
            index = largest

        return max_value

    def heap_sort(self, list1):
        self.heap = []  # Clear any existing data in heap
        self.create_heap(list1)
        sorted_list = []
        try:
            while True:
                sorted_list.insert(0, self.delete_element())
        except EmptyHeapException:
            pass
        return sorted_list


# Test
h = Heap()
print(h.heap_sort([3, 9, 1, 2, 0, -4]))
```
