```python
class DQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.items = [None] * capacity
        self.front = -1
        self.rear = -1

    def is_full(self):
        return (self.rear + 1) % self.capacity == self.front

    def is_empty(self):
        return self.front == -1

    def enqueue_first(self, item):
        if self.is_full():
            print("Queue is full, cannot insert at front")
            return
        if self.is_empty():
            self.front = self.rear = 0
        else:
            self.front = (self.front - 1 + self.capacity) % self.capacity
        self.items[self.front] = item

    def enqueue_last(self, item):
        if self.is_full():
            print("Queue is full, cannot insert at rear")
            return
        if self.is_empty():
            self.front = self.rear = 0
        else:
            self.rear = (self.rear + 1) % self.capacity
        self.items[self.rear] = item

    def dequeue_first(self):
        if self.is_empty():
            print("Queue is empty, cannot delete from front")
            return
        removed = self.items[self.front]
        if self.front == self.rear:  # only one element
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.capacity
        return removed

    def dequeue_last(self):
        if self.is_empty():
            print("Queue is empty, cannot delete from rear")
            return
        removed = self.items[self.rear]
        if self.front == self.rear:  # only one element
            self.front = self.rear = -1
        else:
            self.rear = (self.rear - 1 + self.capacity) % self.capacity
        return removed

    def display(self):
        if self.is_empty():
            print("Queue is empty")
            return
        i = self.front
        print("Queue contents:", end=" ")
        while True:
            print(self.items[i], end=" ")
            if i == self.rear:
                break
            i = (i + 1) % self.capacity
        print()
q1=DQueue(8)
q1.enqueue_first(3)
q1.enqueue_first(7)
q1.enqueue_first(9)
q1.enqueue_first(11)
q1.enqueue_first(15)
q1.display()
```

```python
class CircularQueue:
    def __init__(self, capacity):
        self.queue = [None] * capacity  
        self.front = -1  
        self.rear = -1  
        self.capacity = capacity  
        self.size = 0  

    def enqueue(self, item):
        if self.isFull():
            print("Queue is full! Cannot insert", item)
            return
        if self.size ==0:
            self.front = 0
            self.rear = (self.rear + 1) % self.capacity
        else:
            self.rear = (self.rear + 1) % self.capacity
        self.queue[self.rear] = item
        self.size += 1

    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty! Cannot dequeue.")
            return None
        item = self.queue[self.front]
        self.queue[self.front] = None  
        if self.front == self.rear:
            self.front = -1
            self.rear = -1
        else:
            self.front = (self.front + 1) % self.capacity
        self.size -= 1
        return item

    def isFull(self):
        return self.size == self.capacity

    def isEmpty(self):
        return self.size == 0

    def display(self):
        if self.isEmpty():
            print("Queue is empty!")
            return
        # print(self.queue)
        print("Circular Queue:", end=" ")
        index = self.front
        for _ in range(self.size):
            print(self.queue[index], end=" | ")
            index = (index + 1) % self.capacity
        print()

cq = CircularQueue(5)
cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.enqueue(50)
cq.enqueue(60)  
cq.display()  
cq.dequeue()
cq.dequeue()
cq.display()  
cq.enqueue(60)
cq.enqueue(70)  
cq.display()
```

```python
class Node:
    def __init__(self, data=None, priority=None, next=None):
        self.data = data
        self.priority = priority
        self.next = next

class PriorityQueue:
    def __init__(self):
        self.start = None
        self.item_count = 0

    def enqueue(self, data, priority):
        new_node = Node(data, priority)
        
        # Inserting at front if queue is empty or new node has higher priority (lower value)
        if not self.start or self.start.priority > priority:
            new_node.next = self.start
            self.start = new_node
        else:
            # Traverse to find the correct position based on priority
            temp = self.start
            while temp.next and temp.next.priority <= priority:
                temp = temp.next
            
            # Insert node in its correct position
            new_node.next = temp.next
            temp.next = new_node

        self.item_count += 1

    def dequeue(self):
        if self.start is None:
            print("Priority queue is empty.")
            return None
        item = self.start.data
        self.start = self.start.next
        self.item_count -= 1
        return item

    def display(self):
        if self.start is None:
            print("Priority queue is empty.")
            return
        temp = self.start
        while temp:
            print((temp.data, temp.priority), end=" ")
            temp = temp.next
        print()

# Driver code
p1 = PriorityQueue()
p1.enqueue(9, 0)
p1.enqueue(5, 2)
p1.enqueue(7, 1)
p1.enqueue(3, 0)
p1.display()  # Expected order: (9,0) (3,0) (7,1) (5,2)

```

```python
# Queue implementation using 2 stacks (lists) - Pure Python

class QueueUsingStack:
    def __init__(self):
        # s1: input stack (naye elements yahan push honge)
        # s2: output stack (queue ke front se element pop karne ke liye)
        self.s1 = []
        self.s2 = []

    def enqueue(self, x):
        # Element ko input stack (s1) mein daalo
        self.s1.append(x)

    def dequeue(self):
        # Agar s2 khali hai to s1 ke elements ko reverse karke s2 mein daalo
        if self.is_empty():
            print("Queue khali hai!")
            return None

        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())  # s1 ke elements ko ulta kar ke s2 mein daal rahe hain

        # s2 ka top element (jo sabse purana hai) return karo
        return self.s2.pop()

    def front(self):
        # Queue ka front element dikhata hai bina hataaye
        if self.is_empty():
            print("Queue khali hai!")
            return None

        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())

        return self.s2[-1]

    def is_empty(self):
        # Agar dono stacks khali hain to queue bhi khali hai
        return len(self.s1) == 0 and len(self.s2) == 0
    def display(self):
      if self.is_empty():
        print("Queue khali hai!")
        return

      print("Queue (front to rear):", end=" ")

      # Print elements of s2 from top to bottom (reversed)
      for i in range(len(self.s2) - 1, -1, -1):
        print(self.s2[i], end=" ")

       # Print elements of s1 from bottom to top
      for i in range(len(self.s1)):
        print(self.s1[i], end=" ")

      print()


# ---------- Example Usage ----------
q = QueueUsingStack()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Front element hai:", q.front())   # Output: 10
print("Dequeue kiya:", q.dequeue())      # Output: 10
print("Ab front element hai:", q.front())  # Output: 20
```

```python
# Node class for Doubly Linked List
class Node:
    def __init__(self, data):
        self.data = data      # Data store karega
        self.prev = None      # Previous node ka reference
        self.next = None      # Next node ka reference

# Queue class using Doubly Linked List
class Queue:
    def __init__(self):
        self.front = None     # Queue ka front (dequeue yahin se hoga)
        self.rear = None      # Queue ka rear (enqueue yahan hoga)

    def enqueue(self, value):
        # Naya node banate hain
        new_node = Node(value)

        if self.rear is None:
            # Agar queue empty hai to front aur rear dono new_node honge
            self.front = self.rear = new_node
        else:
            # Rear ke next ko new node se link karo
            self.rear.next = new_node
            new_node.prev = self.rear
            self.rear = new_node  # Rear ko update karo

    def dequeue(self):
        if self.front is None:
            print("Queue khali hai!")
            return None

        # Front node ko hata rahe hain
        value = self.front.data
        self.front = self.front.next

        if self.front is not None:
            self.front.prev = None
        else:
            # Agar ek hi element tha, to rear bhi None kar do
            self.rear = None

        return value

    def peek(self):
        # Front element ko dikhata hai bina hataaye
        if self.front is None:
            print("Queue khali hai!")
            return None
        return self.front.data

    def is_empty(self):
        return self.front is None


# ---------- Example Usage ----------
q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Front element hai:", q.peek())      # Output: 10
print("Dequeue kiya:", q.dequeue())        # Output: 10
print("Ab front element hai:", q.peek())   # Output: 20
```
